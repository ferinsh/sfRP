```python
import pandas as pd
import numpy as np
```

```python
hourly=pd.read_csv("hourly_power_features.csv")
hourly["Datetime"] = pd.to_datetime(hourly["Datetime"])
hourly = hourly.set_index("Datetime")

hourly.head()
```

```python
hour = hourly.index.hour
hourly["hour_sin"] = np.sin(2 * np.pi * hour / 24)
hourly["hour_cos"] = np.cos(2 * np.pi * hour / 24)

dow = hourly.index.dayofweek
hourly["dow_sin"] = np.sin(2 * np.pi * dow / 7)
hourly["dow_cos"] = np.cos(2 * np.pi * dow / 7)

month = hourly.index.month
hourly["month_sin"] = np.sin(2 * np.pi * month / 12)
hourly["month_cos"] = np.cos(2 * np.pi * month / 12)

hourly["is_weekend"] = (hourly.index.dayofweek >= 5).astype(int)
```

```python
hourly.head()
```

```python
target_col = "Global_active_power_mean"

hourly["lag_1"] = hourly[target_col].shift(1)
hourly["lag_24"] = hourly[target_col].shift(24)
hourly["lag_168"] = hourly[target_col].shift(168)

hourly.head()
```

```python
for window in [6, 24, 168]:
    hourly[f"roll_mean_{window}h"] = (
        hourly[target_col]
        .shift(1)
        .rolling(window)
        .mean()
    )

    hourly[f"roll_std_{window}h"] = (
        hourly[target_col]
        .shift(1)
        .rolling(window)
        .std()
    )

hourly[[
    "lag_1", "lag_24", "lag_168",
    "roll_mean_6h", "roll_std_6h",
    "roll_mean_24h", "roll_std_24h",
    "roll_mean_168h", "roll_std_168h",
]].head(200)
```

```python
hourly = hourly.dropna()
hourly.head()
```

```python
hourly.columns
```

```python
target = "Global_active_power_mean"

time_feats = [
    "hour_sin", "hour_cos",
    "dow_sin", "dow_cos",
    "month_sin", "month_cos",
    "is_weekend"
]

electrical_feats = [
    col for col in hourly.columns
    if "Voltage" in col
    or "Sub_metering" in col
    or "Global_intensity" in col
]

lag_feats = [
    "lag_1", "lag_24", "lag_168",
    "roll_mean_6h", "roll_std_6h",
    "roll_mean_24h", "roll_std_24h",
    "roll_mean_168h", "roll_std_168h"
]

F1 = [target]
F2 = F1 + time_feats
F3 = F2 + electrical_feats
F4 = F3 + lag_feats

len(F1), len(F2), len(F3), len(F4)

```

```python
import json

feature_sets = {
    "F1": F1,
    "F2": F2,
    "F3": F3,
    "F4": F4,
}

with open("feature_sets.json", "w") as f:
    json.dump(feature_sets, f, indent=2)

print("Saved feature_sets.json")
```
